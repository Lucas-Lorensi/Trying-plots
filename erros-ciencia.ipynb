```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Number of tests
num_tests = 20
p_values = []

# Perform t-tests
for _ in range(num_tests):
    sample1 = np.random.normal(loc=0, scale=1, size=30)
    sample2 = np.random.normal(loc=0, scale=1, size=30)
    _, p = stats.ttest_ind(sample1, sample2)
    p_values.append(p)

# Select one p-value below 0.05 to highlight
highlight_idx = np.where(np.array(p_values) < 0.05)[0]
if len(highlight_idx) > 0:
    highlight_idx = highlight_idx[0]
else:
    highlight_idx = np.argmin(p_values)  # Highlight the lowest if none below 0.05

# Plot the results
plt.figure(figsize=(10, 5))
plt.bar(range(num_tests), p_values, color=['#d73b3e' if i == highlight_idx else '#0072b1' for i in range(num_tests)],edgecolor='none')
plt.axhline(0.05, color='gray', linestyle='--', label='Significance Threshold (p=0.05)')
plt.xlabel('Teste')
plt.ylabel('p-value')
plt.yticks()
plt.title('Quando vocÃª testa muitas vezes, sempre acha um resultado!')
plt.legend()
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(12)

# Number of tests
num_tests = 20
p_values = []

# Perform t-tests
for _ in range(num_tests):
    sample1 = np.random.normal(loc=0, scale=1, size=30)
    sample2 = np.random.normal(loc=0, scale=1, size=30)
    _, p = stats.ttest_ind(sample1, sample2)
    p_values.append(p)

# Select one p-value below 0.05 to highlight
highlight_idx = np.where(np.array(p_values) < 0.05)[0]
if len(highlight_idx) > 0:
    highlight_idx = highlight_idx[0]
else:
    highlight_idx = np.argmin(p_values)  # Highlight the lowest if none below 0.05

# Plot the results

plt.figure(figsize=(10, 5),facecolor='lightgray')
plt.bar(range(1, num_tests + 1), p_values, color=['#d73b3e' if i == highlight_idx + 1 else '#0072b1' for i in range(1, num_tests + 1)], edgecolor='none')
plt.axhline(0.05, color='gray', linestyle='--', label='Limite de significÃ¢ncia (p=0.05)')
plt.xlabel('Teste')
plt.ylabel('p-value')
plt.title('Quando vocÃª testa muitas vezes, sempre acha um resultado!')
plt.legend(edgecolor='none')

# Adjust x and y ticks
plt.xticks(range(1, num_tests + 1))  # Show all test indices
plt.yticks(np.arange(0, 1.1, 0.1))  # Show p-value intervals of 0.1

# Remove all plot borders and grid lines
plt.gca().set_facecolor('lightgray')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

plt.savefig('erros-ciencia.png')
plt.show()
```
