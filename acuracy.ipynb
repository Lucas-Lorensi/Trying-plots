```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")

# Matriz de confusão A: acurácia alta, mas ignora a classe positiva
conf_matrix_a = np.array([[90, 0],
                          [10, 0]])

# Matriz de confusão B: mesma acurácia, mas detecta classe positiva
conf_matrix_b = np.array([[80, 10],
                          [2, 8]])

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Labels
labels = ["Negativa", "Positiva"]

# Plot da matriz A
sns.heatmap(conf_matrix_a, annot=True, fmt="d", cmap="Blues", cbar=False,
            xticklabels=labels, yticklabels=labels, ax=axes[0])
axes[0].set_title("Modelo A\n(Acurácia = 90%)")
axes[0].set_xlabel("Previsto")
axes[0].set_ylabel("Real")

# Plot da matriz B
sns.heatmap(conf_matrix_b, annot=True, fmt="d", cmap="Greens", cbar=False,
            xticklabels=labels, yticklabels=labels, ax=axes[1])
axes[1].set_title("Modelo B\n(Acurácia = 88%)")
axes[1].set_xlabel("Previsto")
axes[1].set_ylabel("")

# Ajustes finais
plt.suptitle("Acurácia engana: contextos importam", fontsize=14, fontweight="bold")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("accuracy")
plt.show()
```
